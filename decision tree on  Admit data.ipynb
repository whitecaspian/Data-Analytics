```python
#Cost complexity pruning
path = ddc.cost_complexity_pruning_path(xtrain, ytrain)
ccp_alphas = path.ccp_alphas

clf_dts=[]

for ccp_alpha in ccp_alphas:
    a=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    a.fit(xtrain,ytrain)
    clf_dts.append(a)
```

```python
clf_dts
```

```python
# alpha values
ccp_alphas
```

```python
path
```

```python
#sklearn version
import sklearn
sklearn.__version__
```

```python
#finding the best alpha value to remove overfitting of model
train_scores = [a.score(xtrain, ytrain) for a in clf_dts]
test_scores = [a.score(xtest, ytest) for a in clf_dts]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
#same without steppost 
train_scores = [a.score(xtrain, ytrain) for a in clf_dts]
test_scores = [a.score(xtest, ytest) for a in clf_dts]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        )
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        )
ax.legend()
plt.show()
```
