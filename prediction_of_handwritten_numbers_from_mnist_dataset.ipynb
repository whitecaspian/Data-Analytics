```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
mnist=keras.datasets.mnist
```

```python
(xtrain,ytrain),(xtest,ytest)=mnist.load_data()
```

```python
plt.imshow(xtrain[1])
plt.show()
```

```python

plt.imshow(xtrain[100])
plt.show()
```

```python
for i in range(0,10):
  plt.subplots()
  plt.imshow(xtrain[i])
```

```python
xtrain=tf.keras.utils.normalize(xtrain,axis=1)
xtest=tf.keras.utils.normalize(xtest,axis=1)
```

```python
mod=tf.keras.Sequential()
mod.add(tf.keras.layers.Flatten())
mod.add(tf.keras.layers.Dense(128,activation='relu'))
mod.add(tf.keras.layers.Dense(128,activation='sigmoid'))
mod.add(tf.keras.layers.Dense(10,activation='sigmoid'))
mod.compile(optimizer='adam',metrics=['accuracy'],loss='sparse_categorical_crossentropy')
```

```python
mod.fit(xtrain,ytrain,epochs=100)
```

```python
trainpredicted=mod.predict(xtrain)
```

```python
np.argmax(trainpredicted[0])
```

```python
np.argmax(trainpredicted[1])
```

```python
np.argmax(trainpredicted[100])
```

```python
testpredicted=mod.predict(xtest)
```
